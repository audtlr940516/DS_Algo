```python
import numpy as np
from scipy.stats import t

np.random.seed(50) # 같은 결과를 얻기 위해 시드 고정
t_sample = t.rvs(df=3, size=20)
t_sample
```

```python
from itertools import combinations

print(" mean : ", t_sample.mean())
print(" median: ", np.median(t_sample))
print("핫지스-레만 일표본중앙값(왈쉬평균들의 중앙값): ", np.median(np.array(list(combinations(t_sample, 2))).mean(1)))
```

```python
t_mean = []
t_median = []
t_hodges = []

for _ in range(100):
    t_sample = t.rvs(df=3, size=20)
    t_mean.append(t_sample.mean())
    t_median.append(np.median(t_sample))
    t_hodges.append(np.median(np.array(list(combinations(t_sample, 2))).mean(1)))

t_mean = np.array(t_mean)
t_median = np.array(t_median)
t_hodges = np.array(t_hodges)

print(t_mean)
print(t_median)
print(t_hodges)
```

```python
print(f't_mean : {t_mean.var()}')
print(f't_median : {t_median.var()}')
print(f't_hodges : {t_hodges.var()}')
```

핫지스 레만 추정량은 -> 데이터의 모든 쌍 평균을 만든 뒤, 그 중앙값을 가져온 값이다.
이 핫지스 레만 추정량은 outlier 에 더 강하고, median 보다 정보를 더 많이 활용하는 중간 성격의 robust estimator 이다. 
t_mean 과 t_median 과 t_hodges 추정량 결과를 비교해보면, mean 의 분산이 튀었고,
median 과 hodges 의 분산이 작다. ->  mean 값이 이상치에 영향을 많이 받은 결과.
