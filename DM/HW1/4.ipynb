```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
np.random.seed(42)
n = 1000
sample_normal = np.random.normal(loc=0,scale=np.sqrt(2),size=n)
sample_chi2 = np.random.chisquare(df=5,size=n)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(sample_normal,bins=40)
plt.title("N(0,2)")

plt.subplot(1,2,2)
plt.hist(sample_chi2,bins=40)
plt.title("x^2(5)")

def sample_skewness(x):
    n = len(x)
    x_bar = np.mean(x)
    s = np.std(x,ddof=1)
    return np.sum(((x-x_bar)/s)**3 / (n-1))


skew_normal_manual = sample_skewness(sample_normal)
skew_chi2_manual = sample_skewness(sample_chi2)


skew_normal_scipy=stats.skew(sample_normal)
skew_chi2_scipy=stats.skew(sample_chi2)

print("=" * 50)
print("왜도(Skewness) 결과")
print("=" * 50)
print(f"N(0,2) - 수식 : {skew_normal_manual:.4f} , scipy : {skew_normal_scipy:.4f}")
print(f"x^2(5) - 수식 : {skew_chi2_manual:.4f} , scipy : {skew_chi2_scipy:.4f}")


sample_t1 = np.random.standard_t(1,n)
sample_t10 = np.random.standard_t(10,n)
sample_norm = np.random.normal(0,1,n)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.hist(sample_t1, bins=30)
plt.title('t(1)')

plt.subplot(1,3,2)
plt.hist(sample_t10, bins=30)
plt.title('t(10)')

plt.subplot(1,3,3)
plt.hist(sample_norm, bins=30)
plt.title('norm')

plt.show()

def sample_kurtosis(x):
    n = len(x)
    x_bar  = np.mean(x)
    s = np.sqrt(np.sum((x-x_bar)**2) / (n-1))
    return np.sum(((x-x_bar)/s)**4) / (n-1)

kurt_t1 = sample_kurtosis(sample_t1)
kurt_t10 = sample_kurtosis(sample_t10)
kurt_norm = sample_kurtosis(sample_norm)

print(f"t(1)    첨도: {kurt_t1:.4f}")
print(f"t(10)   첨도: {kurt_t10:.4f}")
print(f"N(0, 1) 첨도: {kurt_norm:.4f}")
```
