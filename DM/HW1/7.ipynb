```python
import numpy as np
from scipy import stats

np.random.seed(50)

draws = []
for i in range(10):
    one_try = np.random.choice(range(1,46),size=6,replace=False)
    draws.append(one_try)
print(draws)

```

```python
count = np.zeros(45)
for draw in draws:
    for num in draw:
        count[num - 1] += 1
p_hat = count / 10
print(p_hat)
```

```python
p_values = np.zeros(45)
for i in range(45):
    x = count[i]
    result = stats.binomtest(int(x), n=10, p=6/45, alternative='greater')
    p_values[i] = result.pvalue
p_num = sum(p_values < 0.05)
print(p_num)
```

```python
alpha_bonf = 0.05 / 45           

n_d = np.sum(p_values < alpha_bonf)      
print(f"답: {n_d}개")
```
