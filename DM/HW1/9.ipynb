```python
import numpy as np
import pandas as pd
from scipy import stats

df = pd.read_csv('height_male.csv')
x = df.iloc[:,1].values
n = len(x)
sigma = 6.33

trust = stats.norm.interval(0.95,loc=x.mean(),scale=sigma/np.sqrt(n))
print(f"CLT 95% = ({trust[0]:.4f} , {trust[1]:.4f})")


```

```python
B = 10000   # bootstrap 횟수
n = len(x)

# bootstrap 평균 저장
bootstrap_means = []

for _ in range(B):
    sample = np.random.choice(x, size=n, replace=True)
    bootstrap_means.append(np.mean(sample))

bootstrap_means = np.array(bootstrap_means)

# 95% percentile interval
lower = np.percentile(bootstrap_means, 2.5)
upper = np.percentile(bootstrap_means, 97.5)


print(f"Bootstrap 95% : ({lower:.4f}, {upper:.4f})")
```

CLT 방법과 Bootstrap 방법 두가지는 결과는 거의 유사하다.
n 의 수가 200으로 표본 크기가 충분히 크기 때문에 중심극한정리에 의해서 표본 평균이 정규분포에 근사하다. Bootstrap 은 표본을 가짜 모집단으로 가정하고 사용하여, 표본 평균의 분포를 직접 근사하기 때문에 두 결과는 비슷하게 나오게 된다. 
하지만 CLT 방법의 경우 정규성 가정 + 모표준편차 정보 필요한 반면, Bootstrap의 경우 분포에 대한 가정 없이 사용 가능하기 때문에, data 가 비정규 or 표본 크기 작은 경우에도 사용 가능하다.
