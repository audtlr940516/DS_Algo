```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import scipy.stats as stats

# 데이터
y = np.array([5.0, 7.9, 11.2, 11.7, 8.9, 6.8, 8.4, 8.8, 10.1, 7.5])
x1 = np.array([1.0, 1.1, 1.5, 3.0, 2.4, 1.2, 2.4, 2.9, 1.3, 2.0])
x2 = np.array([0.5, 2.1, 3.0, 1.5, 1.7, 1.2, 0.2, 0.9, 2.3, 1.0])

X = np.column_stack((x1, x2))

# 회귀
model = LinearRegression()
model.fit(X, y)

# 예측값
y_hat = model.predict(X)

# 잔차
residuals = y - y_hat

# QQ plot
stats.probplot(residuals, dist="norm", plot=plt)
plt.title("QQ Plot")
plt.show()
```

QQ Plot 에서 점들이 대체적으로 직선 근처에 위치하고 있다. 따라서 잔차는 정규분포를 따른다고 할 수 있다. 잔차가 정규분포를 따르니 정규성을 만족한다고 볼 수 있다.
