```python
def varience_cal(pi1, pi2, mu1, mu2, var1, var2, n):
    var = (pi1 * var1 + pi2 * var2 + pi1 * (mu1**2) + pi2 * (mu2**2) - (pi1 * mu1 + pi2 * mu2)**2) / n
    return var

print(f'표본 평균 분산 계산 결과 : {varience_cal(0.5, 0.5, 160, 173, 4, 5, 100)}')
```

```python
def calc_var_strat(pi1, pi2, var1, var2, n1, n2):
    var = (pi1**2) * var1 / n1 + (pi2**2) * var2 / n2
    return var

print(f'층화추출 분산 계산 결과 : {calc_var_strat(0.5, 0.5, 4, 5, 50, 50)}')
```

평균의 추정 결과는 층화추출을 이용하여 data 를 처리한 경우, 단순하게 data 를 처리하는 경우보다 더 분산이 작게 나온다.
이는 추정값이 흔들리는 정도가 더 낮다는 것이고, 더 정확한 값을 나타내 준다는 것이다. (신뢰도가 더 높음)
키의 경우, 남녀 사이에 분명한 차이가 존재하기 때문에, 단순임의추출로 뽑은 표본에서는 -> 남녀의 비율이 달라질 가능성이 크고
이로 인해 변동이 커진다. 하지만 층화 추출의 경우, 남녀를 정확히 반으로 나눠(층을 만듬) 뽑기 때문에 비율이 일정하게 되고
변동성이 줄어들게 되어, 분산이 작아지게 된다.
