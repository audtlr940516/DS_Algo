```python
import numpy as np
from scipy import stats

taga = [23.5 ,12.0 ,21.0 ,22.0 ,19.1 ,21.5 ,22.1 ,20.4 ,18.3 ,21.6 ,23.3 ,21.0 ,22.1 ,23.0 ,12.0]
jaga = [17.4 ,20.4 ,20.0 ,20.0 ,18.4 ,18.6 ,18.6 ,15.3 ,16.5 ,18.0 ,16.3 ,18.0 ,12.8 ,15.5 ,18.0]

diff = np.zeros(len(taga))

for i in range(len(taga)):
    diff[i] = taga[i] - jaga[i]

n = len(diff)
diff_mean = np.mean(diff)
diff_sd = np.std(diff,ddof=1)
T = diff_mean / (diff_sd / np.sqrt(n))

result = stats.ttest_rel(taga, jaga)
ci = result.confidence_interval(confidence_level=0.95)

print(f"T 통계량      = {result.statistic:.4f}")
print(f"양측 p-value  = {result.pvalue:.4f}")
print(f"95% 신뢰구간  = ({ci.low:.4f}, {ci.high:.4f})")
print("\n T-검정 실시 결과 p-value가 0.05 보다 높고, 95% 신뢰구간에 0 가 포함되어 귀무가설을 기각할 수 없다.")
```
