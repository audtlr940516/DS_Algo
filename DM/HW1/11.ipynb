```python
import numpy as np
np.random.seed(50)
n=50

S1 = np.random.uniform(-3,3,size=(n,1))
S1[1]

```

```python
y = np.array([1 if x[0]>0 else 0 for x in S1])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    S1, y, test_size=0.5, random_state=50
)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# K-NN 모델 (K=3) 생성 및 학습
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# 테스트 데이터 예측 및 정확도 계산
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f" result : {accuracy}")
```

```python
accuracies = []
dimensions = range(1, 41)

for d in dimensions:

    X = np.random.uniform(-3, 3, (50, d))
    y = np.array([1 if x[0]>0 else 0 for x in X])

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
    knn = KNeighborsClassifier(n_neighbors=3)
    knn.fit(X_train, y_train)

    acc = accuracy_score(y_test, knn.predict(X_test))
    accuracies.append(acc)

accuracies
```

```python
import matplotlib.pyplot as plt

plt.plot(dimensions, accuracies, marker='o')
plt.title('Accuracy vs Dimension')
plt.xlabel('Dimension (d)')
plt.ylabel('Accuracy')
plt.grid(True)
plt.show()
```
