```python
import numpy as np

data = np.array([0.302, 1.383, -0.281,-0.876,-0.759, -0.248, 0.009, 1.019,-0.935, 0.275, 0.413, 0.438,-0.135,-0.091, 0.666, 0.877,-1.135,-1.02,-0.035, 0.353,-1.05, 0.203, 0.43, 0.992,-0.058,-2.26,-0.532, 0.907, 1.3, 1.018])
```

```python
from scipy.stats import ttest_1samp, wilcoxon

print(f't검정 결과 (피밸류 값): {ttest_1samp(data, popmean=0).pvalue}')
print(f'wilcoxon 부호 검정 결과 (피밸류 값): {wilcoxon(data, zero_method="wilcox").pvalue}')
```

두개 검정 결과 모두 귀무가설 기각 불가능

```python
data_add = np.append(data, [20,20,20,20])

print(f't검정 결과 (피밸류 값): {ttest_1samp(data_add, popmean=0).pvalue}')
print(f'wilcoxon 부호 검정 결과 (피밸류 값): {wilcoxon(data_add, zero_method="wilcox").pvalue}')
```

이상치가 없을 경우에는, t검정 와 윌콕슨 부호 검정 결과 모두 귀무가설을 기각시키지 않으며, 정상적으로 추정을 하는 반면, 이상치가 포함된 data set 의 경우, t검정 결과 p-value 가 0.05 이하로 떨어지며 95% 검정을 하는 경우 mean이 0 이라는 귀무가설을 기각시킬 수 있게 된다. 반면 윌콕슨 추정의 경우 p-value 가 감소하기는 하였지만, 그래도 귀무가설을 기각시키지는 않는다. 따라서 윌콕슨 부호 검정은 이상치에 더 robust 하다고 할 수 있다.
