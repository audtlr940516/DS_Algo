```python
import numpy as np
np.random.seed(50)
n=10000

S1 = np.random.uniform(-1,1,size=n)
S2 = np.random.uniform(-1,1,size=n)

S1_sum = np.zeros(n)
S2_sum = np.zeros(n)

S1_sum[0] = S1[0]
S2_sum[0] = S2[0]

for i in range(1,n):
    S1_sum[i] = S1_sum[i-1] + S1[i]
for i in range(1,n):
    S2_sum[i] = S2_sum[i-1] + S2[i]

corr = np.corrcoef(S1_sum,S2_sum)[0,1]
print(f"상관계수 : {corr}")
```
